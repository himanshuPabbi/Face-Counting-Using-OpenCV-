```python
# Import required libraries
import cv2
import numpy as np
import dlib
from google.colab.patches import cv2_imshow

# Load the image from file
img = cv2.imread('/content/2face.jpg')  # Replace with your image path

# Check if the image was loaded successfully
if img is None:
    print("Error: Could not read the image.")
    exit()

# Convert image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Initialize face detector
detector = dlib.get_frontal_face_detector()

# Detect faces in the image
faces = detector(gray)

# Iterate through each face detected
i = 0
for face in faces:
    # Get the coordinates of the face
    x, y = face.left(), face.top()
    x1, y1 = face.right(), face.bottom()

    # Draw rectangle around the face
    cv2.rectangle(img, (x, y), (x1, y1), (0, 255, 0), 2)

    # Increment iterator for each face
    i = i + 1

    # Display face number on top of the rectangle
    cv2.putText(img, 'face num ' + str(i), (x-10, y-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

# Display the image with detected faces
cv2_imshow(img)

# Wait until a key is pressed, then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
```
